```r
library(lattice)
library(grid)
library(gridExtra)
```

```r
s <- function(x){(137*x + 187) %% 256}
```

```r
lcgseq <- rep(NA, 258)
lcgseq[1] <- 1
for(i in 1:257){
    lcgseq[i+1] <- s(lcgseq[i])
}
print(length(unique(lcgseq))) # make sure we got each value at least once.
table(lcgseq) # check that we got each value once, except we should get 1 and 68 twice because the period wraps around
```

```r
triples <- data.frame(lcgseq[1:256], lcgseq[2:257], lcgseq[3:258])
colnames(triples) <- c("d1", "d2", "d3")
head(triples)
```

```r
p1 <- xyplot(d2~d1, data = triples)
```

```r
p2 <- cloud(d3~d1*d2, data = triples)
```
